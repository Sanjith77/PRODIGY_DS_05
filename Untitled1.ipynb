```python
import pandas as pd

# Load your data into final_data
final_data = pd.read_csv("US_Accidents_March23.csv")
```

```python
final_data.isnull().sum()
```

```python
print("State Code: ",final_data.State.unique())
print("Total No. of State in Dataset: ",len(final_data.State.unique()))
```

```python
fig, ax = plt.subplots(figsize=(20, 5))
c = sns.countplot(x='State', data=final_data, order=final_data['State'].value_counts().index)
c.set_title("States with No. of Accidents")
plt.show()
```

```python
final_data['Start_Time'] = pd.to_datetime(final_data['Start_Time'], errors='coerce')
```

```python
# Extract Year from Start_Time column
final_data['Year'] = final_data['Start_Time'].dt.year
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

# Create countplot with hue as Year and legend=False
c = sns.countplot(x="Year", data=final_data, palette="crest", hue="Year", legend=False)

# Annotate the plot for missing data
plt.annotate('Data Not Available', xy=(-0.4, 500000), fontsize=11)

# Set title
c.set_title("No. of Accidents in Year")

# Annotate each bar with the count
for i in ax.patches:
    count = '{:,.0f}'.format(i.get_height())  # Format count with comma
    x = i.get_x() + i.get_width() / 2  # X-position of the label (centered on the bar)
    y = i.get_height() + 10000  # Y-position above the bar
    ax.annotate(count, (x, y), ha='center', fontsize=10)

# Show the plot
plt.show()
```

```python
data_2016 = final_data[final_data['Year'] == 2016]
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

# Create countplot for months in 2016
c = sns.countplot(x="Month", data=data_2016,hue="Month",palette="crest",legend=False)

# Annotate the plot for missing data
plt.annotate('Data Not Available', xy=(2, 50000), fontsize=11)

# Set the title
c.set_title("No. of Accidents in Months of Year 2016")

# Show the plot
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

# Create countplot for months in 2016
c = sns.countplot(x="Month", data=data_2016, orient = 'v', hue="Month", palette="crest",legend=False)

# Annotate the plot for missing data
plt.annotate('Data Not Available', xy=(2, 50000), fontsize=11)

# Set the title
c.set_title("No. of Accidents in Months of Year 2016")

# Show the plot
plt.show()
```

```python
# Plot the accident counts by month for 2016 (not 2020)
fig, ax = plt.subplots(figsize=(10, 5))

# Create countplot for months in 2016 without 'hue' but setting legend=False
c = sns.countplot(x="Month", data=data_2016,  hue="Month", palette="crest")

# Add annotations for the Covid-19 pandemic (if relevant in context)
plt.annotate('Covid-19 Pandemic', xy=(2, 150000), fontsize=12)
plt.annotate("[", xy=(0, 0), xytext=(1.9, 150000), arrowprops={'arrowstyle': '-|>'}, fontsize=12)
plt.annotate("]", xy=(10, 0), xytext=(4.5, 150000), arrowprops={'arrowstyle': '-|>'}, fontsize=12)

# Set title of the plot
c.set_title("No. of Accidents in Months of Year 2016")

# Show the plot
plt.show()
```
